```python
df['TV'] = df['TV'].fillna(df['TV'].mean())
df['Radio'] = df['Radio'].fillna(df['Radio'].mean())
df['Social Media'] = df['Social Media'].fillna(df['Social Media'].mean())
df['Sales'] = df['Sales'].fillna(df['Sales'].mean())
```

```python
plt.figure()
plt.scatter(df['TV'], df['Sales'])
plt.title('TV Advertising vs Sales')
plt.xlabel('TV Budget')
plt.ylabel('Sales')
plt.show()
```

```python
avg_values = df[['TV', 'Radio']].mean()

plt.figure()
avg_values.plot(kind='bar')
plt.title('Average Advertising Spend')
plt.xlabel('Media Type')
plt.ylabel('Average Budget')
plt.show()
```
